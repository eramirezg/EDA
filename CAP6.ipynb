# Correlacin

## Introduccin a la correlacin

Cualquier conjunto de datos que queramos analizar tendr diferentes campos (es decir, columnas) de mltiples observaciones (es decir, variables) que representan diferentes hechos. Las columnas de un conjunto de datos estn, muy probablemente, relacionadas entre s porque se recopilan del mismo evento. Un campo de registro puede o no afectar el valor de otro campo. Para examinar el tipo de relaciones que tienen estas columnas y analizar las causas y efectos entre ellas, tenemos que trabajar para encontrar las dependencias que existen entre las variables. La fuerza de dicha relacin entre dos campos de un conjunto de datos se denomina correlacin, que se representa mediante un valor numrico entre -1 y 1.


En otras palabras, la tcnica estadstica que examina la relacin y explica si, y con qu fuerza, los pares de variables estn relacionados entre s se conoce como correlacin. La correlacin responde a preguntas como cmo cambia una variable con respecto a otra?. Si cambia, en qu grado o fuerza? Adems, si la relacin lineal entre esas variables es lo suficientemente fuerte, entonces podemos hacer predicciones para el comportamiento futuro.

Por ejemplo, la altura y el peso estn relacionados; Es decir, las personas ms altas tienden a ser ms pesadas que las personas ms bajas. Si tenemos una nueva persona que es ms alta que la altura promedio que observamos antes, entonces es ms probable que pese ms que el peso promedio que observamos.

La correlacin nos dice cmo las variables cambian juntas, tanto en la misma direccin como en direcciones opuestas y en que magnitud (es decir, la fuerza) de la relacin. Para encontrar la correlacin, calculamos el coeficiente de correlacin de Pearson, simbolizado por $\rho$ (la letra griega rho). Esto se obtiene dividiendo la covarianza por el producto de las desviaciones estndar de las variables:
$$
\rho_{xy}=\dfrac{Cov[X,Y]}{\sigma_x\sigma_y}
$$

En cuanto a la fuerza de la relacin, el valor de la correlacin entre dos variables, *A* y *B*, vara entre +1 y -1. Si la correlacin es +1, entonces se dice que es una correlacin positiva / lineal perfecta (es decir, la variable *A* es directamente proporcional a la variable *B* en el mismo sentido), mientras que una correlacin de -1 es una correlacin negativa perfecta (es decir, la variable *A* es directamente proporcional a la variable *B* en sentido opuesto). Tenga en cuenta que se supone que los valores ms cercanos a 0 no deben estar correlacionados en absoluto. Si los coeficientes de correlacin estn cerca de 1 en valor absoluto, entonces se dice que las variables estn fuertemente correlacionadas; En comparacin, se dice que aquellos que estn ms cerca de 0.5 estn dbilmente correlacionados.

Echemos un vistazo a algunos ejemplos usando diagramas de dispersin. Los diagramas de dispersin muestran cunto afecta una variable por otra:



Como se muestra en el primer y ltimo grfico, cuanto ms cercana sea la distancia entre los puntos de datos cuando se trazan para hacer una lnea recta, mayor ser la correlacin entre las variables asociadas. Cuanto mayor sea la correlacin entre ellos, ms fuerte ser la relacin entre las variables. Cuanto ms dispersos estn los puntos de datos cuando se graficen (sin crear patrones), menor ser la correlacin entre las dos variables. Aqu, debe observar los siguientes cuatro puntos importantes

- Cuando el grfico de puntos de datos tiene una lnea recta que pasa por el origen hasta los valores *x* e *y*, entonces se dice que las variables tienen una **correlacin positiva**. 
- Cuando los puntos de datos se trazan para generar una lnea que va desde un valor alto en el eje y hasta un valor alto en el eje *x*, se dice que las variables tienen una **correlacin negativa**. 
- Una correlacin perfecta tiene un valor de 1. 
- Una correlacin negativa perfecta tiene un valor de -1.

A una correlacin altamente positiva se le da un valor ms cercano a 1. A una correlacin altamente negativa se le da un valor ms cercano a -1. En el diagrama anterior, +0.8 da una correlacin positiva alta y -0.8 da una correlacin negativa alta. Cuanto ms cerca est el nmero de 0 (en el diagrama, esto es +0.3 y -0.3), ms dbil ser la correlacin.

### Comprender el anlisis univariado

Recuerda las variables con las que trabajamos en el Captulo 4, Estadstica Descriptiva, para las medidas de estadstica descriptiva? All tenamos un conjunto de enteros que iban del 2 al 12. Tal anlisis en un solo tipo de conjunto de datos se denomina **anlisis univariante**.

El anlisis univariado es la forma ms sencilla de analizar datos. Significa que nuestros datos tienen un solo tipo de variable y que realizamos anlisis sobre ella. El propsito principal del anlisis univariado es tomar datos, resumir esos datos y encontrar patrones entre los valores. No trata de causas o relaciones entre los valores. Varias tcnicas que describen los patrones encontrados en los datos univariados incluyen la tendencia central (es decir, la media, la moda y la mediana) y la dispersin (es decir, el rango, la varianza, los cuartiles mximo y mnimo (incluido el rango intercuartlico) y la desviacin estndar).

### Anlisis multivariante

El anlisis multivariante es el anlisis de tres o ms variables. Esto nos permite observar las correlaciones (es decir, cmo cambia una variable con respecto a otra) e intentar hacer predicciones para el comportamiento futuro con mayor precisin que con el anlisis bivariado.

Inicialmente, exploramos la visualizacin del anlisis univariado y el anlisis bivariado; Asimismo, visualicemos el concepto de anlisis multivariante.

Una forma comn grficar datos multivariados es hacer un diagrama de dispersin de matriz, conocido como grfico de pares. Una grfica de matriz o grfica de pares muestra cada par de variables trazadas entre s. El grfico de pares nos permite ver tanto la distribucin de variables individuales como las relaciones entre dos variables:

Podemos usar la funcin `scatter_matrix()` del paquete `pandas.tools.plotting` o la funcin `seaborn.pairplot()` del paquete seaborn para hacer esto:


```python
# grfico de pares con regresin de tipo de grfico
sns.pairplot(df,height=1.5,vars=['normalized-losses','price','horsepower'],
             kind="reg")
plt.show()
```

Como se muestra en el diagrama anterior, el histograma en diagonal nos permite ilustrar la distribucin de una sola variable. Los grficos de regresin en los tringulos superior e inferior demuestran la relacin entre dos variables. El grfico del medio en la primera fila muestra el grfico de regresin; Esto representa que no hay correlacin entre las prdidas normalizadas y el precio de los automviles. En comparacin, el grfico de regresin media en la fila inferior ilustra que existe una gran correlacin entre el precio y los caballos de fuerza.

```python
 #pair plot (matrix scatterplot) of few columns 
sns.set(style="ticks", color_codes=True)
sns.pairplot(df,height=1.5,vars=['symboling','normalized-losses','wheel-base'], 
             hue="drive-wheels")
plt.show()
```

Como se discuti anteriormente, el anlisis de correlacin es una tcnica eficiente para averiguar si alguna de las variables en un conjunto de datos multivariado est correlacionada. Para calcular el coeficiente de correlacin lineal (Pearson) para un par de variables, podemos usar la funcin `dataframe.corr(method ='pearson')` del paquete pandas y la funcin `pearsonr()` del paquete `scipy.stats`:

Por ejemplo, para calcular el coeficiente de correlacin para el precio y los caballos de fuerza, use lo siguiente:

```python
from scipy import stats

corr = stats.pearsonr(df["price"], df["horsepower"])
print("p-value:\t", corr[1])
print("cor:\t\t", corr[0])
```

Aqu la correlacin de estas dos variables es 0,7579, que est cerca de +1, por lo que podemos asegurarnos de que el precio y la potencia estn correlacionados de forma muy positiva.

### La paradoja de Simpson

Por lo general, las decisiones que tomamos a partir de nuestro conjunto de datos estn influenciadas por la salida de las medidas estadsticas que les aplicamos. Esos resultados nos informan sobre el tipo de correlacin y las visualizaciones bsicas del conjunto de datos. Sin embargo, a veces, las decisiones difieren cuando segregamos los datos en grupos y aplicamos medidas estadsticas, o cuando los agregamos y luego aplicamos medidas estadsticas. Este tipo de comportamiento anmalo en los resultados del mismo conjunto de datos generalmente se denomina paradoja de Simpson. En pocas palabras, la paradoja de Simpson es la diferencia que aparece en una tendencia de anlisis cuando un conjunto de datos se analiza en dos situaciones diferentes: primero, cuando los datos se separan en grupos y, segundo, cuando los datos se agregan.

La tabla anterior presenta la tasa de recomendacin de dos consolas de juegos diferentes: PS4 y Xbox One por hombres y mujeres, tanto individualmente como combinadas.
