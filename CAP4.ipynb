### Función de distribución

Para entender el concepto de la función de distribución, es esencial entender el concepto de una función continua. Entonces, ¿qué queremos decir cuando nos referimos a una función continua? Básicamente, una función continua es cualquier función que no tiene ningún cambio inesperado en el valor. Estos cambios abruptos o inesperados se conocen como discontinuidades. Por ejemplo, considere la siguiente función cúbica:
$$ f(x)=x^3+x^2-5x+3$$
Si graficamos esta función, verá que no hay saltos ni agujeros en la serie de valores. Por lo tanto, esta función es continua:



Tratemos de entender qué es la **función de densidad de probabilidad (PDF)**. El PDF se puede definir en términos de una función continua, en otras palabras, para cualquier función continua, el PDF es la probabilidad de que la variable tenga el valor de x. 

Ahora, si la función está asociada con variables aleatorias discretas en lugar de variables aleatorias continuas, entonces la función se conoce como una **función de masa de probabilidad (PMF)**. 

**La distribución de probabilidad** o **función de probabilidad** de una variable aleatoria discreta es una lista de probabilidades vinculadas a cada uno de sus valores alcanzables. Supongamos que una variable aleatoria, *A*, toma todos los valores en un intervalo de números reales. Entonces, la probabilidad de que *A* esté en la lista de resultados *Z*, *P(Z)*, es el área por encima de *Z* y bajo una curva que describe una función *p(a)* que satisface las siguientes condiciones:

1. La curva no puede tener valores negativos (p(a) > 0 para todo a).
2. El área total bajo la curva es siempre igual a 1.

Tales curvas se conocen como curvas de densidad. Las distribuciones de probabilidad continua incluyen distribución normal, distribución exponencial, distribución uniforme, distribución gamma, distribución de Poisson y distribución binomial.

## Distribución uniforme

La función de distribución de probabilidad uniforme de cualquier distribución uniforme continua viene dada por la siguiente ecuación:
$$
f(x)=
\left\{
  \begin{array}{cl}
    \dfrac{1}{b-a}, & \hbox{para $a\leq x\leq b$;} \\
    0, & \hbox{$x<a$ o $x>b$.}
  \end{array}
\right.
$$
Tracemos el gráfico para una distribución uniforme utilizando las bibliotecas de Python, seaborn y matplotlib. En primer lugar, vamos a importar las bibliotecas importantes necesarias para generar el gráfico:

```python
# Importar librerias
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
import matplotlib.pyplot as plt
from IPython.display import Math, Latex
from IPython.core.display import Image
import seaborn as sns

sns.set(color_codes=True)
sns.set(rc={'figure.figsize':(10,6)})
```

Ahora, generemos una distribución uniforme:

```python
# Distribución Uniforme
from scipy.stats import uniform

number = 10000
start = 20
width = 25

uniform_data = uniform.rvs(size=number, loc=start, scale=width)
axis = sns.histplot(uniform_data, bins=100, kde=True, color='skyblue')
axis.set(xlabel='Distribución Uniforme', ylabel='Frecuencia')
plt.show()
```

La función `uniform` se utiliza para generar una variable continua uniforme entre la ubicación de inicio dada (loc) y el ancho de los argumentos (escala). Los argumentos de tamaño especifican el número de variables aleatorias tomadas en consideración. El gráfico ilustra el hecho de que el conjunto de datos está distribuido uniformemente.


## Distribución normal

La distribución normal, o distribución gaussiana, es una función que distribuye la lista de variables aleatorias en un gráfico que tiene forma de campana simétrica. Una distribución normal tiene una curva de densidad que es simétrica sobre su media, con su propagación típicamente definida por su desviación estándar. Tiene dos parámetros: la media y la desviación estándar. El hecho de que la distribución normal se base principalmente en el teorema del límite central lo hace relevante. Si el tamaño de todas las muestras posibles en una población es $n$, y la media es $\mu$ y la varianza $\sigma^2$, entonces la distribución se aproxima a una distribución normal. Matemáticamente, se da de la siguiente manera:
$$
f(x)=\dfrac{1}{\sqrt{2\pi}\sigma}\exp^{-\frac{(x-\mu)^2}{2\sigma^2}},\quad X\sim N(\mu,\sigma^2)
$$
Veamos cómo podemos dibujar una ilustración para la distribución normal usando la biblioteca de estadísticas de Python:

```python
# Distribución Normal
from scipy.stats import norm

normal_data = norm.rvs(size=90000,loc=20,scale=30)
axis = sns.histplot(normal_data, bins=100, kde=True, color='skyblue')
axis.set(xlabel='Distribución Normal ', ylabel='Frecuencia')
plt.show()
```

Podemos obtener un gráfico de distribución normal usando los módulos `scipy.stats` por el método `norm.rvs()`. Permite que el argumento `loc` establezca la media de la distribución, el argumento de escala `scale` establezca la desviación estándar y, finalmente, el argumento `size` indique el número de muestras aleatorias.

## Distribución Gamma

La PDF para la gamma es de la forma
$$
f(x,a)=\dfrac{x^{a-1}e^{-x}}{\Gamma(a)}
$$
Para $x\geq 0,a>0$. Aqui $\Gamma(a)$ es la función gamma

La distribución Gamma algunas veces es parametrizada con dos parámetros, con PDF de la forma
$$
f(x,\alpha,\lambda)=\dfrac{\lambda^a x^{a-1}e^{-\lambda x}}{\Gamma(a)}
$$
Note que esta parametrización es equivalente a la anterior, con la escala igual a $1/\lambda$.

```python
# Distribution Gamma
from scipy.stats import gamma

gamma_data = gamma.rvs(a=5, size=10000)
axis = sns.histplot(gamma_data, kde=True, bins=100, color='skyblue')
axis.set(xlabel='Ejemplo de Distribución Gamma', ylabel='Frecuencia')
plt.show()
```

## Distribución exponencial

Un proceso en el que algunos eventos ocurren de forma continua e independiente a una tasa promedio constante se conoce como proceso de punto de Poisson. La distribución exponencial describe el tiempo entre eventos en tal proceso puntual de Poisson, y la función de densidad de probabilidad de la distribución exponencial se da de la siguiente manera:
$$
f(x;\lambda)=\left\{
  \begin{array}{cl}
    \lambda e^{-\lambda x}, & x\geq 0; \\
    0, & x<0.
  \end{array}
\right.
$$
Note que está distribución es un caso particular de la Gamma, basta con tomar $a=1$.

```python
# Distribution Exponencial
from scipy.stats import expon

expon_data = expon.rvs(scale=1,loc=0,size=1000)
axis = sns.histplot(expon_data, kde=True, bins=100, color='skyblue')
axis.set(xlabel='Distribucion Exponencial', ylabel='Frecuencia')
plt.show()
```

El gráfico ilustra la función exponencial decreciente. La curva está disminuyendo sobre el eje x.

## Distribución de Poisson

La variable aleatoria de Poisson cuenta el número de éxitos en $n$ ensayos independientes de Bernoulli en el límite cuando $n\longrightarrow \infty$ y $p\longrightarrow\infty$ donde la probabilidad de éxito en cada ensayo es $p$ y $np=\lambda\geq 0$ es una constante. Puede usarse para aproximar la variable aleatoria binomial o por sí solo para contar el número de eventos que ocurren en el intervalo $[0,t]$. La PMF es de la forma:
$$
p(x;\lambda)=e^{-\lambda}\dfrac{\lambda^x}{x!}
$$


```python
# Distribución de Poisson
from scipy.stats import poisson

poisson_data = poisson.rvs(mu=2, size=10000)
axis = sns.histplot(poisson_data, bins=30, kde=False, color='red')
axis.set(xlabel='Distribución de Poisson', ylabel='Frecuencia')
plt.show()
```

El método `poisson.rvs()` del módulo `scipy.stats` toma $\lambda$ como el argumentos de la media de la distribución para generar el gráfico.

## Distribución binomial

La distribución binomial, como su nombre indica, tiene solo dos resultados posibles, éxito o fracaso. No es necesario que los resultados sean igualmente probables y cada ensayo es independiente del otro.

```python
from scipy.stats import binom

binomial_data = binom.rvs(n=10, p=0.8,size=10000)

axis = sns.histplot(binomial_data, kde=False, color='red')
axis.set(xlabel='Distribución Binomial', ylabel='Frecuencia')
plt.show()
```

El método `binom.rvs()` del módulo `scipy.stats` toma $n$ como el número de argumentos de prueba, y $p$ como la probabilidad de éxito como parámetros de forma para generar el gráfico.

## Función de distribución acumulativa

La función de distribución acumulativa (CDF) es la probabilidad de que la variable tome un valor menor o igual que x. Matemáticamente, está escrito de la siguiente manera:
$$
F(x)=P[X\leq x]=\alpha
$$
Para una variable aleatoria continua, La CDF proporciona el área bajo la PDF, que va desde menos infinito hasta x.

## Estadística descriptiva

La estadística descriptiva se ocupa de la formulación de resúmenes simples de datos para que puedan entenderse claramente. Los resúmenes de datos pueden ser representaciones numéricas o visualizaciones con gráficos simples para una mayor comprensión. Por lo general, estos resúmenes ayudan en la fase inicial del análisis estadístico. Hay dos tipos de estadística descriptiva:

1. Medidas de tendencia central
2. Medidas de variabilidad
3. Medidas de forma

Las medidas de tendencia central incluyen media, mediana y moda, las medidas de variabilidad incluyen desviación estándar (o varianza), los valores mínimos y máximos de las variables y la medidas de forma incluyen los coeficientes de curtosis y de asimetría. Vamos a discutir estas categorías en la siguiente sección.


## Medidas de tendencia central

La medida de tendencia central tiende a describir el valor promedio o medio de los conjuntos de datos que se supone que proporciona un resumen óptimo de todo el conjunto de mediciones. Este valor es un número que de alguna manera es central para el conjunto. Las medidas más comunes para analizar la frecuencia de distribución de los datos son la media, la mediana y el modo.

### Media/promedio

La media, o promedio, es un número alrededor del cual se distribuyen las variables continuas observadas. Este número estima el valor de todo el conjunto de datos. Matemáticamente, es el resultado de la división de la suma de números por el número de enteros en el conjunto de datos.

Sea $x$ un conjunto de enteros

$
x=(12,2,3,5,8,9,6,4,2)
$

Por lo tanto, el valor medio de x se puede calcular de la siguiente manera:
$$
Mean(x)=\dfrac{12+2+3+5+8+9+6+4+2}{9}=5.66
$$
A continuación, veamos la mediana.

### Mediana

Dado un conjunto de datos que se ordena en orden ascendente o descendente, la mediana divide los datos en dos partes. La fórmula general para calcular la mediana es la siguiente:
$$
\text{posición de la mediana}=\frac{(n+1)}{2}\text{ésima observación}
$$
Aquí, $n$ es el número de elementos en los datos. Los pasos involucrados en el cálculo de la mediana son los siguientes:

1. Ordene los números en orden ascendente o descendente. 
2. Si $n$ es impar, busque el *(n+1)/2*-ésimo término. El valor correspondiente a este término es la mediana.
3. Si $n$ es par, encuentre el *(n+1)/2*-ésimo término. El valor mediano es el promedio de los números a cada lado de la posición mediana.

Para un conjunto de enteros como $x$, debemos organizarlos en orden ascendente y luego seleccionar el entero medio. 

$x$ en orden ascendente $= (2,2,3,4,5,6,8,9,12)$.

Aquí, la mediana es 5.



### Moda

La moda es el número que más veces se repite en el conjunto de datos. Resulta ser el valor con la frecuencia más alta en el conjunto de datos. En el dataset $x$ en el ejemplo mediano, la moda es 2 porque ocurre dos veces en el conjunto.

## Medidas de dispersión

El segundo tipo de estadística descriptiva es la medida de dispersión, también conocida como medida de variabilidad. Se utiliza para describir la variabilidad en un conjunto de datos, que puede ser una muestra o población. Por lo general, se utiliza junto con una medida de tendencia central, para proporcionar una descripción general de un conjunto de datos. Una medida de dispersión/variabilidad/dispersión nos da una idea de qué tan bien la tendencia central representa los datos. Si estamos analizando el conjunto de datos de cerca, a veces, la media/promedio podría no ser la mejor representación de los datos porque variará cuando haya grandes variaciones entre los datos. En tal caso, una medida de dispersión representará la variabilidad en un conjunto de datos con mucha más precisión.

Múltiples técnicas proporcionan las medidas de dispersión en nuestro conjunto de datos. Algunos métodos comúnmente utilizados son la desviación estándar (o varianza), los valores mínimos y máximos de las variables, el rango, la curtosis y la asimetría.

### Desviación estándar

En lenguaje sencillo, la desviación estándar es el promedio/media de la diferencia entre cada valor del conjunto de datos con su promedio/media; es decir, cómo se distribuyen los datos de la media. Si la desviación estándar del conjunto de datos es baja, entonces los puntos de datos tienden a estar cerca de la media del conjunto de datos, de lo contrario, los puntos de datos se distribuyen en un rango más amplio de valores.

 Diferentes bibliotecas de Python tienen funciones para obtener la desviación estándar del conjunto de datos. La biblioteca NumPy tiene la función `numpy.std(dataset)`. La biblioteca de estadísticas tiene la función `statistics.stdev(dataset)`. Usando la biblioteca pandas, calculamos la desviación estándar en nuestro marco de datos df usando la función `std()`:

```python
# desviación estándar de conjunto de datos utilizando la función std()
std_dev =df.std(numeric_only=True)
print(std_dev)
```

```python
# desviación estándar de una columna especifica
sv_height=df.loc[:,"height"].std()
print(sv_height)
```

### Medidas de varianza

La varianza es el cuadrado de la desviación estándar.

Diferentes bibliotecas de Python tienen funciones para obtener la varianza del conjunto de datos. La biblioteca NumPy tiene la función `numpy.var(dataset)`. La biblioteca **statistic** tiene la función `statistics.variance(dataset)`. Usando la biblioteca pandas, calculamos la varianza en nuestro dataframe df usando la función `var()`:


```python
# Varianza del conjunto de datos utilizando la función var()
variance=df.var(numeric_only=True)
print(variance)
```

```python
# varianza de una columna especifica
var_height=df.loc[:,"height"].var()
print(var_height)
```

Es esencial tener en cuenta las siguientes observaciones del fragmento de código proporcionado aquí:

- Es importante tener en cuenta que `df.var()` calculará la varianza en el dataframa dado a través de la columna de forma predeterminada. Además, podemos especificar *axis=0* para indicar que necesitamos calcular la varianza por columna o por fila.
- Al especificar `df.var(axis=1)` se calculará la varianza en fila en el dataframe dado.
- Finalmente, también es posible calcular la varianza en cualquier columna en particular especificando la ubicación. Por ejemplo, `df.loc[:,"height"].var()` calcula la varianza en la columna `height` en el conjunto de datos anterior.


## Medidas de forma

las medidas de forma sirven para determinar cómo es una distribución sin necesidad de representarla gráficamente.

En estadística, se distinguen dos medidas de forma:

- **Asimetría (skew)**: indica si una distribución es simétrica o asimétrica.
- **Curtosis (kurt)**: indica si una distribución es escarpada o achatada.

### Asimetría

En teoría de la probabilidad y estadística, la asimetría es una medida de la asimetría de la variable en el conjunto de datos sobre su media. El valor de asimetría puede ser positivo o negativo, o indefinido. El valor de asimetría nos dice si los datos están sesgados o son simétricos. Aquí hay una ilustración de un conjunto de datos sesgado positivamente, datos simétricos y algunos datos sesgados negativamente:



Tenga en cuenta las siguientes observaciones del diagrama anterior:

- El gráfico en el lado derecho tiene una cola que es más larga que la cola en el lado derecho. Esto indica que la distribución de los datos está sesgada hacia la izquierda. Si selecciona cualquier punto en la cola más larga de la izquierda, la media es menor que el moda. Esta condición se conoce como **asimetría negativa**.

- El gráfico en el lado izquierdo tiene una cola que es más larga en el lado derecho. Si selecciona cualquier punto de la cola derecha, el valor medio es mayor que el moda. Esta condición se conoce como **asimetría positiva**. 

- El gráfico en el medio tiene una cola derecha que es la misma que la cola izquierda. Esta condición se conoce como condición **simétrica**.

### Kurtosis

Ya hemos discutido la distribución normal, el gráfico en forma de campana. Básicamente, la curtosis es una medida estadística que ilustra cuán fuertemente difieren las colas de distribución de las de una distribución normal. Esta técnica puede identificar si una distribución dada contiene valores extremos.

La curtosis, a diferencia de la asimetría, no se trata de la picosidad o la planitud. Es la medida de la presencia atípica en una distribución dada. Tanto la curtosis alta como la baja son un indicador de que los datos necesitan más investigación. Cuanto mayor es la curtosis, mayores son los valores atípicos.

### Tipos de curtosis

Hay tres tipos de curtosis: mesokurtica, leptokurtica y platykurtic. Veamos estos uno por uno:

- **Mesocurtica**: Si cualquier conjunto de datos sigue una distribución normal, sigue una distribución mesokurtic. Tiene curtosis alrededor de 0.

- **Lapticurtica**: En este caso, la distribución tiene curtosis mayor que 3 y las colas gordas indican que la distribución produce más valores atípicos.

- **Platicurtica**: En este caso, la distribución tiene curtosis negativa y las colas son muy delgadas en comparación con la distribución normal.

Diferentes bibliotecas de Python tienen funciones para obtener la curtosis del conjunto de datos. La biblioteca SciPy tiene la función `scipy.stats.kurtosis(dataset)`. Usando la biblioteca pandas, calculamos la curtosis de nuestro dataframe df usando la función `kurt()`:

```python
# Curtosis de los datos usando la función skew ()
kurtosis =df.kurt(numeric_only=True)
print(kurtosis)

# Curtosis de una columna específica
sk_height=df.loc[:,"height"].kurt()
print(sk_height)
```

### Calculando percentiles

Los percentiles miden el porcentaje de valores en cualquier conjunto de datos que se encuentran por debajo de un cierto valor. Para calcular los percentiles, debemos asegurarnos de que nuestra lista esté ordenada. Un ejemplo sería si dijeras que el percentil 80 de datos es 130: entonces, ¿qué significa eso? Bueno, simplemente significa que el 80% de los valores se encuentran por debajo de 130. Usaremos la siguiente fórmula para esto:

La formula para calcular los percentiles de $X=\dfrac{\text{Número de valores menores que  }X}{\text{Número total de observaciones}}*100$

Suponga que tenemos los datos dados: 1, 2, 2, 3, 4, 5, 6, 7, 7, 8, 9, 10. Entonces el valor percentil de 4 = (4/12) * 100 = 33.33%.

Esto simplemente significa que el 33,33% de los datos es inferior a 4.
Ahora, calculemos el percentil de la columna de altura del mismo dataframe que hemos estado usando hasta ahora:

```python
# calcular el percentil 50 de height en el conjunto de datos
height = df["height"]
percentile = np.percentile(height, 50,)
print(percentile)
```

La fórmula anterior es muy simple. Pero, ¿ve algún patrón con las medidas de las tendencias centrales? ¿Cuál sería el percentil 50? Esto corresponde a la *mediana*.


**Cuartiles**

Divide el conjunto de datos en cuatro puntos iguales.

Primer cuartil = percentil 25

Segundo cuartil = percentil 50 (mediana)

Tercer cuartil = percentil 75

Según el cuartil, existe otra medida llamada rango intercuartílico que también mide la variabilidad en el conjunto de datos. Se define como:

IQR = Q3 - Q1

IQR no se ve afectado por la presencia de valores atípicos. 

```python
price = df.price.sort_values()
Q1 = np.percentile(price, 25)
Q2 = np.percentile(price, 50)
Q3 = np.percentile(price, 75)

IQR = Q3 - Q1
IQR
```
