## Gráfico de líneas

¿Recuerdas qué es una variable continua y qué es una variable discreta? Si no es así eche un vistazo rápido al *Capítulo 1, Fundamentos del análisis exploratorio de datos*. Volviendo al tema principal, un gráfico de líneas se utiliza para ilustrar la relación entre dos o más variables continuas. continuas.

Vamos a utilizar la biblioteca `matplotlib`.

````python
# Importar
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
````

````python
# Datos de ejemplo
años = [ 2018 , 2019 , 2020 , 2021 ] 
valores = [ 100 , 120 , 150 , 200 ] 
plt.plot(años, valores) 
plt.xlabel('Año') 
plt.ylabel ('Valores') 
plt.title('Ejemplo de gráfico de líneas') 
plt.xticks(años) 
plt.show()
````

Cambia el estilo de la linea las opciones

"-" o "solid": línea sólida (es la opción por defecto)

"--" o "dashed": línea discontinua

"-." o "dashdot": línea que alterna guiones y puntos

":" o "dotted": línea de puntos

"None", "" o "": no muestra nada.

````python
plt.plot(años, valores, linestyle='--') 
plt.xlabel('Año') 
plt.ylabel ('Valores') 
plt.title('Ejemplo de gráfico de líneas') 
plt.xticks(años) 
plt.show()
````

Para el grosor de la linea 

````python
plt.plot(años, valores, linewidth=5) 
plt.xlabel('Año') 
plt.ylabel ('Valores') 
plt.title('Ejemplo de gráfico de líneas') 
plt.xticks(años) 
plt.show()
````

Los marcadores [https://matplotlib.org/api/markers_api.html#module-matplotlib.markers] 

````python
plt.plot(años, valores, marker ="p", markersize=7) 
plt.xlabel('Año') 
plt.ylabel ('Valores') 
plt.title('Ejemplo de gráfico de líneas') 
plt.xticks(años) 
plt.show()
````

Este parámetro controla el color de la gráfica y acepta gran cantidad de formatos distintos:

Uno de los siguientes nombres: 'blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'olive' o 'cyan'
Uno de los siguientes caracteres representando colores: 'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w' (por ejemplo, "b" representa "blue", azul)
Tuplas de tres o cuatro valores entre 0 y 1 representando colores en formato RGB o RGBA, por ejemplo: (0.1, 0.3, 0.2)
Una cadena de texto representando un número hexadecimal correspondiente a un color en formato RGB o RGBA, por ejemplo #A055CC, o #99CC3377
Un color en formato X11/CSS4

Cualquier nombre de color web [https://en.wikipedia.org/wiki/Web_colors] En dicha página de la wikipedia podemos ver el listado de nombres colores, el propio color, su representación hexadecimal RGB y su representación decimal RGB:

````python
plt.plot(años, valores, color='MediumBlue', linewidth=3) 
plt.xlabel('Año') 
plt.ylabel ('Valores') 
plt.title('Ejemplo de gráfico de líneas') 
plt.xticks(años) 
plt.show()
````

El parámetro `label` asigna a una gráfica una etiqueta que será mostrada en la leyenda (si activamos ésta). Por ejemplo, creemos tres gráficas con datos tipo "random walk" como los usados en los ejemplos anteriores y asignémosles tres etiquetas cualesquiera:

````python
import numpy as np

plt.plot(np.random.randn(50).cumsum(), label="Gráfica 1")
plt.plot(np.random.randn(50).cumsum(), label="Gráfica 2")
plt.plot(np.random.randn(50).cumsum(), label="Gráfica 3")
plt.legend()
plt.show()
````

En lugar de especificar el color de la gráfica, el marcador a usar y el estilo de línea mediante los parámetros color, marker y linestyle que hemos visto, podemos añadir tras los parámetros **x** e **y** de la función `plot` (o tan solo tras el parámetro **y**, si no hemos especificado las **x**) el parámetro [fmt] con el mismo objetivo. Este parámetro es una cadena de texto en la que podemos incluir (en cualquier orden) un carácter que indique el color de la línea, otro que indique el marcador a usar y uno o dos más para indicar el estilo de la línea. No es necesario añadir los tres datos. Por ejemplo, en la siguiente gráfica se fija el color verde (indicado por el carácter "g"):

````python
y=np.random.randn(50).cumsum()
plt.plot(y, "g")
plt.show()
````

En el siguiente ejemplo, indicamos como marcador el círculo (letra "o") y línea discontinua (caracteres "--"):

````python
y=np.random.randn(50).cumsum()
fig, ax =plt.subplots()
g1=ax.plot(y, 'ro--')
plt.show()
````

Si se utiliza este parámetro [fmt] es posible mostrar dos o más gráficas en el mismo conjunto de ejes ejecutando la función plot una sola vez. En el siguiente ejemplo tenemos dos conjuntos de valores y y los incluimos como argumentos de la función plot separados por el argumento [fmt]:

````python
y1=np.random.randn(50).cumsum()
y2=np.random.randn(50).cumsum()
plt.plot(y1, '-g', y2, '-.*b')
plt.show()
````

## La función plot estilo OO

Todas las herramientas vistas están disponibles en el estilo OO (Object Oriented). Para la generación de la figura y de uno (o más) conjuntos de ejes podemos recurrir a la función matplotlib.pyplot.subplots [https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html] que, como ya hemos visto anteriormente, crea estos objetos y devuelve una referencia a la figura y al conjunto de ejes (si solo se ha creado uno) o a una lista de referencias a los conjuntos de ejes (si se han creado dos o más):
```
fig, ax =plt.subplots()
```
Una vez tenemos la referencia al conjunto de ejes **ax** (supongamos que es solo uno), podemos crear una gráfica ejecutando el método plot sobre ax:
```
g = ax.plot(np.random.randn(50).cumsum())
```
El método `plot` devuelve una lista de objetos Line2D que representan la gráfica, objetos sobre los que es posible trabajar fijando sus propiedades. Por este motivo, almacenamos el resultado del método en una variable a la que llamamos g. En este caso la lista de objetos está formada por un único elemento:

````python
fig, ax =plt.subplots()
g= ax.plot(np.random.randn(50).cumsum())
print(len(g))
````

Podemos, por lo tanto, dar formato al objeto g[0] ejecutando los método equivalentes a algunos de los argumentos de la función plot vistos:

color → set_color

linestyle → set_linestyle

linewidth → set_linewidth

marker → set_marker

markersize → set_markersize

````python
fig, ax =plt.subplots()
g= ax.plot(np.random.randn(50).cumsum())
g[0].set_color("RoyalBlue")
g[0].set_linestyle('--')
g[0].set_linewidth(3)
g[0].set_marker('o')
g[0].set_markersize(8)
plt.show()
````

Para probar los estilos de gráficas ("default", "steps"...) vamos a crear dos gráficas, una junto a otra, con los dos estilos mencionados. El método equivalente al atributo visto es:

drawstyle → set_drawstyle

````python
fig, ax=plt.subplots(1,2)
y=np.random.randn(50).cumsum()

g=ax[0].plot(y)
# g[0].set_drawstyle('default') #opción por defecto

g=ax[1].plot(y)
g[0].set_drawstyle("steps")

plt.show()
````

 la ejecución de la función plt.subplots(1, 2) crea una matriz de conjuntos de ejes formada por una fila y dos columnas (es decir, dos conjuntos de ejes).

Por último, nos queda probar el equivalente del atributo label:

label → set_label

````python
fig, ax=plt.subplots()

g1=ax.plot(np.random.randn(50).cumsum())
g1[0].set_label('Gráfica 1')

g2=ax.plot(np.random.randn(50).cumsum())
g2[0].set_label("Gráfica 2")

plt.legend()
plt.show()
````

El parámetros figsize, que nos permite especificar el tamaño de la figura en pulgadas, indicando en primer lugar el ancho y, a continuación, el alto. Estos valores pueden pasarse como una lista o como una tupla. El valor por defecto es [6.4, 4.8].

````python
fig=plt.figure(figsize=[8,6])
y=np.random.randn(50).cumsum()
plt.plot(y)
plt.show()
````

La función `suptitle` nos permite establecer un título para la figura -no para las gráficas que estén contenidas en ella-:

````python
y1=np.random.randn(50).cumsum()
y2=np.random.randn(50).cumsum()

fig, ax=plt.subplots(1,2)
fig.set_size_inches(10,4)
plt.suptitle('Ingresos y gastos 2019', fontsize=16)
ax[0].plot(y1)
ax[1].plot(y2)
plt.show()
````

El método add_subplot asociado a una figura añade un conjunto de ejes a la misma pero suponiéndolo en una matriz de columnas y filas (de conjuntos de ejes) y creándolo en la posición indicada de dicha matriz.

````python
y1=np.random.randn(50).cumsum()
y2=np.random.randn(50).cumsum()
y3=np.random.randn(50).cumsum()

fig=plt.figure()
fig.add_subplot(2,3,5)
plt.plot(y1)
fig.add_subplot(2,3,1)
plt.plot(y2, color='red')
fig.add_subplot(2,3,3)
plt.plot(y3, color='green')
plt.show()
````

En el ejemplo anterior partimos de 3 conjuntos de datos (y1, y2 e y3) a mostrar en sendas gráficas y, para cada uno de ellos creamos un conjunto de ejes con el método add_subplot. En los tres casos se indica una matriz de 2 filas y 3 columnas. El primer conjunto de ejes creado -con la gráfica de color azul- se sitúa en la posición 2 (las posiciones se cuentan a partir de 1, de izquierda a derecha y de arriba abajo). El segundo conjunto de ejes -con la gráfica de color rojo- se sitúa en la posición 1, es decir, en la esquina superior izquierda de la matriz. Y el tercer conjunto de ejes -con la gráfica de color verde- se sitúa en la posición 6, es decir, en la esquina inferior derecha. El resto de espacios en la matriz quedan sin ocupar.

Podemos añadir un **título** con la función `title`. Esta función acepta como primer argumento el texto del título y otros parámetros que representan propiedades del texto, por ejemplo:

alpha: transparencia del texto

backgroundcolor: color de fondo

color: color del texto

fontfamily: familia de la fuente a usar

fontname: nombre de la fuente a usar

fontsize: tamaño de la fuente

fontstretch: ancho de la fuente

fontstyle: estilo de la fuente ('normal', 'italic', 'oblique')

fontvariant: variante de la fuente ('normal', 'small-caps')

fontweight: "peso" de la fuente ( 'ultralight', 'light', 'normal', 'regular', 'bold', etc.)

horizontalalignment: alineación horizontal del texto ('center', 'right', 'left')

linespacing: espacio entre líneas

position: posición x e y del título

rotation: ángulo de rotación del texto (en grados)

verticalalignment: alineación vertical del texto ('center', 'top', 'bottom', 'baseline', 'center_baseline')

x: posición x del texto

y: posición y del texto

Los parámetros `position`, **x** e **y** toman normalmente valores entre 0 y 1, correspondiendo el 0 al extremo izquierdo del espacio ocupado por el conjunto de ejes y el 1 al extremo derecho (o a los extremos inferior y superior, si nos referimos a la y). Pueden tomar también valores negativos o superiores a 1, pero en este caso el texto se mostrará fuera de los límites de los ejes.

Todos estos parámetros son extremadamente útiles pues podremos aplicarlos no solo al título, sino a las etiquetas que demos a los ejes, a las etiquetas de las marcas de los ejes, etc.

Veamos algunos ejemplos. En el siguiente simplemente ejecutamos la función plt.title pasando como argumento el texto a mostrar:

````python
y =np.random.randn(100).cumsum()

fig, ax=plt.subplots()
plt.plot(y)
plt.title('Ventas 2019')
plt.show()
````

En el siguiente ejemplo especificamos una fuente tamaño 20 y alineamos el texto a la derecha (es decir, el extremo derecho del texto se muestra en el centro del área disponible):

````python
fig, ax=plt.subplots()
plt.plot(y)
plt.title('Ventas 2019', 
          fontsize=20, 
          horizontalalignment='right')
plt.show()
````

Por último, en el siguiente ejemplo configuramos la posición del texto (es decir, su posición x e y en la figura), su tamaño, color y peso:

````python
fig, ax=plt.subplots()
plt.plot(y)
plt.title('Ventas 2019',
          position=(0.75, 0.85),
          fontsize=18,
          color='SteelBlue',
          fontweight='bold')
plt.show()
````

Todos los parámetros relativos al texto tienen un método equivalente en el estilo "orientado a objetos". Estos métodos se aplican a la variable que represente el conjunto de ejes. Por ejemplo, el equivalente al código de la anterior imagen es:

````python
fig, ax=plt.subplots()
ax.plot(y)
title=ax.set_title('Ventas 2019')
title.set_position(xy=[0.75, 0.85])
title.set_fontsize(18)
title.set_color('SteelBlue')
title.set_fontweight('bold')
plt.show()
````

Es posible asignar etiquetas a los ejes x e y con las funciones `xlabel` y `ylabel`. Ambas funciones aceptan un primer argumento con el texto y aceptan también todos los atributos del texto que hemos visto para el título:

````python
y =np.random.randn(100).cumsum()

fig, ax=plt.subplots()
plt.plot(y)
plt.title('Ventas 2019')
plt.xlabel('Año en curso')
plt.ylabel('Millones de dólares')
plt.show()
````

Los métodos del conjunto de ejes equivalentes en el estilo OO son set_xlabel y set_ylabel, ambos devolviendo una referencia al objeto texto creado:

````python
fig, ax=plt.subplots()
ax.plot(y)
title=ax.set_title('Ventas 2019')
xlabel=ax.set_xlabel('Año en curso')
ylabel=ax.set_ylabel('Millones de dólares')
ax.grid()
plt.show()
````

El método `legend` de un conjunto de ejes muestra la leyenda en el gráfico. Vimos en su momento que uno de los parámetros que podemos pasar a la función plot es label. Esta etiqueta es la que se mostrará en la leyenda representando a la gráfica. Para ver cómo funciona, partamos de dos conjuntos de datos a representar en sendas gráficas, y mostremos éstas con su label correspondiente:

````python
y1 =np.random.randn(100).cumsum()
y2 =np.random.randn(100).cumsum()

fig, ax=plt.subplots()
ax.plot(y1, label="Ene")
ax.plot(y2, label="Feb")
ax.legend()
plt.show()
````

Estas etiquetas se pueden sobreescribir al llamar a la función o método legend si pasamos los nuevos valores como argumentos:

````python
fig, ax=plt.subplots()
ax.plot(y1)
ax.plot(y2)
ax.legend(['Enero','Febrero'])
plt.show()
````

Resulta relativamente fácil crear figuras geométricas en nuestras gráficas. Todo lo que necesitamos hacer es importar el objeto adecuado de la sublibrería `matplotlib.patches` (Rectangle, Circle, Ellipse, Polygon, etc.), crear una instancia del objeto, configurarla y, por último, añadir la figura a nuestro conjunto de ejes con el método ax.add_patch.

Para ver un ejemplo, comencemos importando los objetos mencionados y generando los valores y a mostrar en la gráfica:

````python
from matplotlib.patches import Rectangle, Circle, Ellipse, Polygon
````

````python
y=np.random.randn(100).cumsum()
````

A continuación, comencemos mostrando un rectángulo en nuestra gráfica. Para crear una instancia de este objeto, tenemos que pasar una tupla con las coordenadas **x** e **y** de la esquina inferior izquierda, la anchura y la altura del rectángulo. Adicionalmente podemos configurar atributos del rectángulo, muchos de los cuales ya son familiares: alpha, facecolor, edgecolor, etc.:

````python
fig, ax=plt.subplots()
ax.plot(y)
rect=Rectangle((45,-8),13,3, 
               linewidth=2,
               alpha=0.2,
               facecolor="Moccasin",
               edgecolor="red")
ax.add_patch(rect)
plt.show()
````

La configuración del círculo es semejante, sustituyendo el ancho y el alto por el radio:

````python
fig, ax=plt.subplots()
ax.plot(y)
circle=Circle((40,0),2, 
               linewidth=2,
               alpha=0.2,
               facecolor="SaddleBrown",
               edgecolor="red")
ax.add_patch(circle)
plt.show()
````

Vemos que si los ejes x e y no tienen la misma escala, el círculo se muestra como una elipse.

Probemos, por último, con el polígono. Para configurarlo hay que pasar, en una lista, tuplas con las posiciones de los puntos que lo definen. De esta forma, para crear, por ejemplo, un triángulo, habría que pasar tres tuplas, una para cada esquina:

````python
fig, ax=plt.subplots()
ax.plot(y)
polygon=Polygon([[35,0],[40,2],[55,-3]], 
               linewidth=2,
               alpha=0.2,
               facecolor="SaddleBrown",
               edgecolor="red")
ax.add_patch(polygon)
plt.show()
````

# Gráfico de barras (Bar Chart) 

Este es uno de los tipos de visualización más comunes que casi todo el mundo debe haber encontrado. Las barras pueden dibujarse horizontal o verticalmente para representar **variables categóricas**.

Los gráficos de barras se utilizan con frecuencia para distinguir objetos entre distintas colecciones con el fin de seguir las variaciones a lo largo del tiempo. En la mayoría de los casos, los gráficos de barras son muy convenientes cuando los cambios son grandes.

````python
x= np.arange(7) # coordenadas de x
y= np.random.randint(1, 20, 7) #  Altura de la barra
fig, ax= plt.subplots()
ax.bar(x,y)
plt.show()
````

Es posible indicar una coordenada y para la base de una columna utilizando el parámetro **bottom**. Esto nos permite mostrar una gráfica de barras apiladas utilizando el mismo conjunto de **x** generando varias gráficas en el mismo conjunto de ejes, cada una teniendo como inicio (*bottom*) la suma de las anteriores:

````python
x= np.arange(7) 
y1= np.random.randint(1, 10, 7)
y2= np.random.randint(1, 10, 7)
y3= np.random.randint(1, 10, 7)
fig, ax= plt.subplots()
ax.bar(x,y1, label='Producto A')
ax.bar(x,y2, bottom=y1, label="Producto B")
ax.bar(x,y3, bottom=y1+y2, label="Producto C")
ax.legend()
plt.show()
````

Otro de los parámetros disponibles es **width**, que controla el ancho de las barras, permitiéndonos jugar con su apariencia:

````python
x= np.arange(7) # coordenadas de x
y= np.random.randint(1, 20, 7) #  Altura de la barra
fig, ax= plt.subplots()
ax.bar(x,y, width=1)
plt.show()
````

````python
fig, ax= plt.subplots()
ax.bar(x,y, width=0.5)
plt.show()
````

La función `barh` crea el gráfico de barras horizontales

````python
fig, ax= plt.subplots()
ax.barh(x,y, height=0.8)
plt.show()
````

````python
x= np.arange(7) 
y1= np.random.randint(1, 10, 7)
y2= np.random.randint(1, 10, 7)
y3= np.random.randint(1, 10, 7)
fig, ax= plt.subplots()
ax.barh(x,y1, label='Producto A')
ax.barh(x,y2, left=y1, label="Producto B")
ax.barh(x,y3, left=y1+y2, label="Producto c")
ax.legend(loc=(1.1, 0.8))
plt.show()
````

# Gráfico de dispersión

Los diagramas de dispersión utilizan un **sistema de coordenadas cartesianas** para mostrar los valores de dos variables típicas de un conjunto de datos.

¿Cuándo utilizar un gráfico de dispersión? Los gráficos de dispersión pueden construirse en las siguientes dos situaciones:

- Cuando una variable continua depende de otra variable, que está bajo control del observador 
- Cuando ambas variables continuas son independientes

Existen dos conceptos importantes: **variable independiente** y **variable dependiente**. En la modelización estadística o matemática, los valores de las variables dependientes dependen de los valores de las variables independientes. La variable dependiente es la variable de resultado que se estudia. Las variables independientes también se denominan regresores. El mensaje para llevar aquí es que los diagramas de dispersión se utilizan cuando necesitamos mostrar la relación entre dos variables, y por ello a veces se denominan diagramas de correlación. Profundizaremos en más detalles sobre la correlación en el Capítulo 6, Correlación.

Tanto si eres un científico de datos experto como si eres un estudiante de informática principiante, sin duda te habrás encontrado alguna vez con algún tipo de gráfico de dispersión. A pesar de su simplicidad, estos gráficos son potentes herramientas de visualización. Las razones principales son que tienen mucho poder de representación, muchas opciones de diseño, y son lo suficientemente flexibles como para representar un gráfico de formas atractivas.

Algunos ejemplos en los que los gráficos de dispersión son adecuados son los siguientes

- Algunas investigaciones han establecido con éxito que la cantidad de horas de sueño requeridas por una persona depende de la edad de la persona.

El conjunto de datos contiene 50 ejemplos de cada una de las tres especies diferentes de Iris, llamadas setosa, virginica y versicolor. Cada ejemplo tiene cuatro atributos diferentes: `petal_length`, `petal_width`, `sepal_length` y `sepal_width`.

El conjunto de datos se puede cargar de varias maneras.

````python
import seaborn as sns
# Cargue el conjunto de datos de Iris
data = sns.load_dataset('iris')
````

la función `shape` no indica el tamaño de los datos leidos

````python
data.shape
````

la función `head` se utiliza para visualizar las primeras filas del data.frame

Confirmamos que se trata de 150 registros y que es en la última columna ("species") donde se almacena la especie de la flor.

Vamos a cruzar las variables "petal_length" y "petal_width", información que cargaremos en las variables x e y. Pero vamos a usar también el parámetro **c** de la función `plt.scatter` que permite indicar cuál es el color de cada uno de los puntos. Es decir, necesitamos una estructura tipo array de 150 elementos que indiquen el color de cada punto, color que representará la especie de cada muestra. Para ello, vamos a crear un diccionario que asocie cada tipo de especie con un color, y vamos a "mapear" la columna "species" con este diccionario:

````python
set(data['species']) # Especies distintas
````

````python
colors={'setosa': 'Crimson', 'versicolor':'RoyalBlue', 'virginica':'DarkSeaGreen'}
species_color=data.species.map(colors)
````

Ahora ya podemos pasar a la función `plt.scatter` las dos primeras columnas del dataframe y la recién calculada estructura "species_color" (una serie pandas):

````python
fig,ax = plt.subplots()
ax.scatter(data.petal_length, data.petal_width, color= species_color)
plt.show()
````

Es apreciable la distancia que separa el cluster representado por los puntos rojos (flores "setosa") de los otros dos que se muestran más mezclados.

Si queremos añadir una leyenda nos encontramos con un pequeño problema pues, si recuerdas, la leyenda toma las etiquetas de las diferentes gráficas que se incluyesen en los ejes: si incluíamos tres gráficas (cada una con sus datos), asignábamos a cada una una etiqueta, y era esta etiqueta la mostrada en la leyenda. Esto, trasladado a nuestro ejemplo, significa que si quisiéramos que la leyenda distinguiera entre 'setosa', 'versicolor' y 'virginica', tendríamos que crear tres gráficos de dispersión, uno para cada tipo de flor, y dar a cada uno una etiqueta diferente. Y solo entonces podríamos mostrar la leyenda. Aun cuando resulta un poco más complejo de lo que cabría esperar, podemos hacerlo sin mayor problema. El código es el siguiente:

````python
fig, ax = plt.subplots()
for species in set(data.species):
    ax.scatter(
        data.petal_length[data.species == species],
        data.petal_width[data.species == species],
        s = 30,
        c = colors[species],
        label = species
    )
plt.xlabel('Longitud del pétalo')
plt.ylabel('Ancho de pétalo')
plt.legend()
plt.show()
````

Como ves, tras crear la figura y el conjunto de ejes, recorremos en bucle los tipos de especies y, para cada uno, añadimos un diagrama de dispersión al que pasamos los datos del dataframe *data* que se corresponden solo con la especie que estemos considerando. El parámetro **s** controla el tamaño de los puntos, y seguimos asignando el mismo color que antes a cada especie (usando el mismo diccionario que teníamos).

# Gráfico de burbujas

Un gráfico de burbujas es una manifestación del gráfico de dispersión en el que cada punto de datos del gráfico se muestra como una burbuja. Cada burbuja se puede ilustrar con un color, tamaño y apariencia diferente.

Sigamos usando el conjunto de datos Iris para obtener un gráfico de burbujas. Aquí, lo importante a tener en cuenta es que seguiremos usando el método `plt.scatter` para dibujar un gráfico de burbujas:


````python
# Craer gráfico de burbujas
plt.scatter(data.petal_length, data.petal_width,
            s=30*data.petal_length*data.petal_width, 
            c=species_color,
            alpha=0.4 )
plt.xlabel('Longitud del pétalo')
plt.ylabel('Ancho de pétalo')
plt.show()
````

# Gráfico polar

¿Recuerdas el eje polar de la clase de matemáticas? Bueno, un gráfico polar es un diagrama que se traza en un eje polar. Sus coordenadas son el ángulo y el radio, a diferencia del sistema cartesiano de coordenadas x e y. A veces, también se le conoce como gráfico de telaraña. Veamos cómo podemos trazar un ejemplo de un gráfico polar.

Primero, vamos a crear el conjunto de datos:

````python
# Supongamos que tiene 5 cursos en su año académico. 
curso = ["Programación en C", "Métodos numéricos", "Sistema operativo",
         "DBMS", "Redes informáticas"]

# Y planeaste obtener las siguientes calificaciones en cada materia
NotaEsperada = [90, 95, 92, 68, 68, 90]

# Sin embargo, después de tu examen final, esta es la calificación que obtuviste.
NotaActual = [75, 89, 89, 80, 80, 75]
````

Ahora que el conjunto de datos está listo, intentemos crear un gráfico polar. El primer paso importante es inicializar el diagrama de araña. Esto se puede hacer configurando el tamaño de la figura y la proyección polar. Esto debería estar claro por ahora. Tenga en cuenta que en el conjunto de datos anterior, la lista de calificaciones contiene una entrada adicional. Esto se debe a que es una trama circular y necesitamos conectar el primer punto y el último punto para formar un flujo circular. Por lo tanto, copiamos la primera entrada de cada lista y la agregamos a la lista. En los datos anteriores, las entradas 90 y 75 son las primeras entradas de la lista respectivamente. Veamos cada paso:

````python
# 1. Importar bibliotecas requeridas
import matplotlib.pyplot as plt
import numpy as np
````

````python
# 2. Inicializar la gráfica por tamaño de figura y proyección polar
plt.figure(figsize =(10, 6))
plt.subplot(polar = True)

# 3. Prepare el conjunto de datos.
# 4. Configurar theta

theta = np.linspace(0, 2*np.pi, len(NotaActual))

# 5. Haga que las líneas de la cuadrícula se alineen con cada uno de 
# los nombres de los sujetos.
lines, labels = plt.thetagrids(range(0, 360, int(360/len(curso))),(curso))

# 6. Usamos el método plot para trazar el gráfico. Y llene el área debajo de él. 
plt.plot(theta, NotaActual)
plt.fill(theta, NotaActual, 'b', alpha=0.2)

# 7. Ahora, graficamos la calificación esperada obtenida
plt.plot(theta, NotaEsperada)

# 8. Finalmente, agregamos una leyenda y un título agradable y 
# comprensible a la gráfica. 
plt.legend(labels=('Calificación actual','Calificación esperada'))
plt.title("Calificaciones actual frente a esperada por curso")

# 9. Por último, mostramos la gráfica en la pantalla. 
plt.show()
````

Como se ilustra en el resultado anterior, las calificaciones planificadas y reales por curso se pueden distinguir fácilmente. La leyenda aclara qué línea indica las calificaciones planificadas (la línea azul en la captura de pantalla) y qué línea indica las calificaciones reales (la línea naranja en la captura de pantalla). Esto da una indicación clara de la diferencia entre las calificaciones previstas y reales de un estudiante para el público objetivo.

# Histograma

Los gráficos de histograma se utilizan para representar la distribución de cualquier variable continua. Estos tipos de gráficos son muy populares en el análisis estadístico.

Considere los siguientes casos de uso. Una encuesta creada en sesiones de formación profesional de desarrolladores tuvo 100 participantes. Tenían varios años de experiencia en programación de Python que iban de 0 a 20.

Importemos las bibliotecas requeridas y creemos el conjunto de datos:

````python
import numpy as np
import matplotlib.pyplot as plt

# 1. Crear conjunto de datos
añosdeexperiencia = np.array([10, 16, 14,  5, 10, 11, 16, 14,  3, 14, 13, 19,  2,  5,  7,  3, 20,
       11, 11, 14,  2, 20, 15, 11,  1, 15, 15, 15,  2,  9, 18,  1, 17, 18,
       13,  9, 20, 13, 17, 13, 15, 17, 10,  2, 11,  8,  5, 19,  2,  4,  9,
       17, 16, 13, 18,  5,  7, 18, 15, 20,  2,  7,  0,  4, 14,  1, 14, 18,
        8, 11, 12,  2,  9,  7, 11,  2,  6, 15,  2, 14, 13,  4,  6, 15,  3,
        6, 10,  2, 11,  0, 18,  0, 13, 16, 18,  5, 14,  7, 14, 18])
añosdeexperiencia
````

Para trazar el gráfico de histograma, ejecute los siguientes pasos:

````python
plt.figure(figsize = (10,6))

# 2. Trazar la distribución de la experiencia del grupo
nbins = 12
plt.hist(añosdeexperiencia, bins=nbins)

# 3. Agregue etiquetas al eje y al título
plt.xlabel("Años de experiencia con la programación Python")
plt.ylabel("Frecuencia")
plt.title("Distribución de la experiencia de programación Python en la sesión de formación profesional")

# 4. Dibuje una línea vertical verde en el gráfico en la experiencia promedio:
plt.axvline(x=añosdeexperiencia.mean(), linewidth=3, color = 'g') 

# 5. Mostrar el gráfico
plt.show()
````

Podemos decir que la experiencia promedio de los participantes ronda los 10 años. ¿Podemos mejorar el gráfico para una mejor legibilidad? ¿Qué tal si tratamos de trazar el porcentaje de la suma de todas las entradas en añosdeexperiencia? Además de eso, también podemos trazar una distribución normal utilizando la media y la desviación estándar de estos datos para ver el patrón de distribución. En pocas palabras, la distribución normal también se conoce como distribución gaussiana. El término indica una distribución de probabilidad que es simétrica con respecto a la media, lo que ilustra que los datos cercanos al promedio (media) son más frecuentes que los datos alejados de la media.

Para trazar la distribución, podemos agregar un parámetro de densidad = 1 en la función `plot.hist`. Repasemos el código. Tenga en cuenta que hay cambios en los pasos 1, 4, 5 y 6. El resto del código es el mismo que el del ejemplo anterior:

````python
plt.figure(figsize = (10,6))

# 2. Trazar la distribución de la experiencia del grupo
nbins = 15

n, bins, patches = plt.hist(añosdeexperiencia, bins=nbins, density=1)

# 3. Agregue etiquetas al eje y al título
plt.xlabel("Años de experiencia con la programación Python")
plt.ylabel("Frecuencia")
plt.title("Distribución de la experiencia de programación Python en la sesión de formación profesional")

# 4. Dibuje una línea vertical verde en el gráfico en la experiencia promedio:
plt.axvline(x=añosdeexperiencia.mean(), linewidth=3, color = 'g') 

# 5. Calcule la media y la desviación estándar del conjunto de datos.
mu = añosdeexperiencia.mean()
sigma = añosdeexperiencia.std()

# 6. Adición de una línea de mejor ajuste para la distribución normal.
y = ((1 /(np.sqrt(2*np.pi)*sigma))*np.exp(-0.5*(1/sigma*(bins-mu))**2))

# 7. Graficar la distribución normal
plt.plot(bins, y, '--')

# 8. Mostrar la gráfica
plt.show()

````

El gráfico anterior ilustra claramente que no sigue una distribución normal. Hay muchas barras verticales que están por encima y por debajo de la curva de mejor ajuste para una distribución normal. Quizás se esté preguntando de dónde sacamos la fórmula para calcular el paso 6 en el código anterior. Bueno, hay una pequeña teoría involucrada aquí. Cuando mencionamos la distribución normal, podemos calcular la función de densidad de probabilidad usando la función de distribución gaussiana dada por '((1 / (np.sqrt(2 * np.pi) * sigma)) *
np.exp(-0.5 * (1 / sigma * (bins - mu))**2'.

# Gráfico de piruletas

Se puede usar un gráfico de paleta para mostrar la clasificación en los datos. Es similar a un gráfico de barras ordenadas.

Consideremos el conjunto de datos carDF.

````python
# 1. Leer el conjunto de datos

carDF = pd.read_csv('cardata.csv')

# 2. Agrupe por fabricante y tome el kilometraje promedio
processedDF = carDF[['cty','manufacturer']].groupby('manufacturer').apply(lambda x: x.mean())

# 3. Ordene los valores por cty y restablezca el índice
processedDF.sort_values('cty', inplace=True)
processedDF.reset_index(inplace=True)

# 4. Trace el gráfico
fig, ax = plt.subplots(figsize=(16,10), dpi= 80)
ax.vlines(x=processedDF.index, ymin=0, ymax=processedDF.cty, 
          color='firebrick', alpha=0.7, linewidth=2)
ax.scatter(x=processedDF.index, y=processedDF.cty, s=75, 
           color='firebrick', alpha=0.7)

# 5. Anotar título
ax.set_title('Gráfico de piruletas para el kilometraje en carretera utilizando un conjunto de datos de automóviles', 
             fontdict={'size':22})

# 6. Anotar etiquetas y xticks, ylim
ax.set_ylabel('Millas Por Galón')
ax.set_xticks(processedDF.index)
ax.set_xticklabels(processedDF.manufacturer.str.upper(), 
                   rotation=65, fontdict={'horizontalalignment': 'right', 'size':12})
ax.set_ylim(0, 30)

# 7. Escribe los valores en el gráfico.
for row in processedDF.itertuples():
    ax.text(row.Index, row.cty+.5, s=round(row.cty, 2), 
            horizontalalignment= 'center', 
            verticalalignment='bottom', fontsize=14)

# 8. Mostrar el gráfico en la pantalla
plt.show()
````

Habiendo visto el resultado anterior, ahora sabe por qué se llama gráfico de piruleta, ¿no es así? La línea y el círculo en la parte superior brindan una buena ilustración de los diferentes tipos de automóviles y su consumo asociado de millas por galón. Ahora, los datos tienen más sentido, ¿no?
